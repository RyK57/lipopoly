```python
import nibabel as nib
import matplotlib.pyplot as plt


# Get the voxel data as a NumPy array
data = nifti_img.get_fdata()

# Define the range of slices to display
start_slice = 0
end_slice = 11

# Create a subplot grid
num_slices = end_slice - start_slice + 1
cols = 3  # Number of columns in the grid
rows = (num_slices // cols) + 1  # Calculate the number of rows needed

# Create a figure and iterate through the slices
fig = plt.figure(figsize=(12, 8))

for i, slice_index in enumerate(range(start_slice, end_slice + 1)):
    # Create a subplot for each slice
    ax = fig.add_subplot(rows, cols, i + 1)
    
    # Extract the slice
    slice_data = data[:, :, slice_index]
    
    # Display the slice
    ax.imshow(slice_data, cmap='gray')
    ax.set_title(f'Slice {slice_index}')
    ax.axis('off')

# Adjust subplot layout and display
plt.tight_layout()
plt.show()
```

```python
import os
import shutil
import random

# Define the source directories
withlps_dir = r'C:\Users\rithv\OneDrive\Desktop\cnn_rnn_models\lipopoly\2dimgwithlps'
withoutlps_dir = r'C:\Users\rithv\OneDrive\Desktop\cnn_rnn_models\lipopoly\2dimgwithoutlps'
destination_dir = r'C:\Users\rithv\OneDrive\Desktop\cnn_rnn_models\lipopoly\implem'

# Define the split ratio (e.g., 80% for training, 20% for testing)
split_ratio = 0.8

# Create training and testing directories in the destination folder
train_dir = os.path.join(destination_dir, 'train')
test_dir = os.path.join(destination_dir, 'test')
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Function to copy images from source to destination
def copy_images(source_folder, destination_folder, num_samples):
    # List all subject folders in the source directory
    subject_folders = os.listdir(source_folder)
    
    # Randomly shuffle the subject folders
    random.shuffle(subject_folders)
    
    # Calculate the number of subjects for training and testing
    num_train_subjects = int(num_samples * split_ratio)
    num_test_subjects = num_samples - num_train_subjects
    
    # Copy subject folders to the appropriate destination
    for i, subject_folder in enumerate(subject_folders):
        source_subject_path = os.path.join(source_folder, subject_folder)
        if i < num_train_subjects:
            destination_subject_path = os.path.join(train_dir, subject_folder)
        else:
            destination_subject_path = os.path.join(test_dir, subject_folder)
        shutil.copytree(source_subject_path, destination_subject_path)

# Count the number of subject folders in each source directory
num_withlps_subjects = len(os.listdir(withlps_dir))
num_withoutlps_subjects = len(os.listdir(withoutlps_dir))

# Copy images for subjects with LPS
copy_images(withlps_dir, destination_dir, num_withlps_subjects)

# Copy images for subjects without LPS
copy_images(withoutlps_dir, destination_dir, num_withoutlps_subjects)

# Print a summary
print(f"Copied {num_withlps_subjects} subjects with LPS to training and testing.")
print(f"Copied {num_withoutlps_subjects} subjects without LPS to training and testing.")
```

```python
import os
import shutil

# Define the source directories
train_dir = r'C:\Users\rithv\OneDrive\Desktop\cnn_rnn_models\lipopoly\implem\train'
test_dir = r'C:\Users\rithv\OneDrive\Desktop\cnn_rnn_models\lipopoly\implem\test'

# Function to organize folders based on their names
def organize_folders(source_folder, destination_folder):
    # List all subject folders in the source directory
    subject_folders = os.listdir(source_folder)
    
    # Organize subject folders into "with LPS" and "without LPS" categories
    for subject_folder in subject_folders:
        source_subject_path = os.path.join(source_folder, subject_folder)
        if subject_folder.startswith("ro"):
            destination_subject_path = os.path.join(destination_folder, "with_lps", subject_folder)
        elif subject_folder.startswith("so"):
            destination_subject_path = os.path.join(destination_folder, "without_lps", subject_folder)
        else:
            # Handle folders with different naming conventions or others as needed
            destination_subject_path = os.path.join(destination_folder, "other", subject_folder)
        
        shutil.move(source_subject_path, destination_subject_path)

# Organize folders in the test directory
organize_folders(test_dir, test_dir)

# Organize folders in the train directory
organize_folders(train_dir, train_dir)

# Print a summary
print("Folders in test and train directories have been organized by LPS status.")
```
